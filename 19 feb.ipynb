````python
Q1. What is Flask Framework? What are the advantages of Flask Framework?
Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in
Jupyter Notebook.
Q3. What is App routing in Flask? Why do we use app routes?
Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”
route to show the following details:
Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-9999
Attach the screenshot of the output in Jupyter Notebook.
Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.


**Q1. What is Flask Framework? What are the advantages of Flask Framework?**
Flask is a lightweight and versatile web framework for Python. It is designed to make it easy to build web applications quickly and efficiently. Flask provides the necessary tools and libraries to handle routing, rendering templates, managing sessions, and interacting with databases. It follows the WSGI (Web Server Gateway Interface) standard and is often used for developing small to medium-sized web applications.

**Advantages of Flask Framework:**
1. **Minimalistic:** Flask has a simple and minimal core, allowing developers to add only the components they need.
2. **Flexibility:** Flask gives developers the freedom to choose libraries and tools for different functionalities, promoting a modular approach.
3. **Easy to Learn:** Flask's simplicity makes it beginner-friendly, making it a great choice for those new to web development.
4. **Extensive Documentation:** Flask has comprehensive and well-structured documentation, making it easy to find solutions and understand concepts.
5. **Large Community:** Flask has a large community of developers, which means finding solutions, plugins, and resources is easier.
6. **RESTful Support:** Flask is often used to build RESTful APIs due to its simplicity and ease of integration.
7. **Built-in Development Server:** Flask includes a built-in development server, enabling developers to quickly test their applications.
8. **Widely Used:** Flask is widely adopted and used for various web application projects, making it a reliable choice.
9. **Scalability:** While it's more lightweight than some other frameworks, Flask can be used to build scalable applications.

**Q2. Create a simple Flask application to display ‘Hello World!!’.**
Here's a simple Flask application code snippet to display "Hello World!!":

```python
from flask import Flask

app = Flask(__name__)

@app.route('/')
def hello_world():
    return 'Hello World!!'

if __name__ == '__main__':
    app.run()
```

Unfortunately, I can't display images or screenshots in this text-based environment, such as Jupyter Notebook. However, you can run the provided Flask code in your Jupyter Notebook or any Python environment to see the output.

**Q3. What is App routing in Flask? Why do we use app routes?**
App routing in Flask involves mapping URLs to specific functions or view handlers. In Flask, routes are defined using the `@app.route()` decorator. When a user requests a specific URL, Flask's routing system directs the request to the appropriate view function.

We use app routes to create different pages or views within a web application. Each route corresponds to a specific URL, and the associated view function defines what content should be displayed when that URL is accessed. App routes help organize the structure of the application, making it easier to manage and maintain.

**Q4. Create “/welcome” and “/” routes in Flask.**
Here's the Flask code to create the specified routes:

```python
from flask import Flask

app = Flask(__name__)

@app.route('/welcome')
def welcome():
    return 'Welcome to ABC Corporation'

@app.route('/')
def company_details():
    return '''Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-9999'''

if __name__ == '__main__':
    app.run()
```

Unfortunately, I can't display images or screenshots in this text-based environment, such as Jupyter Notebook. However, you can run the provided Flask code in your Jupyter Notebook or any Python environment to see the output.

**Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.**
In Flask, the `url_for()` function is used for URL building. It generates a URL for a given view function and its arguments. Here's an example:

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'Home Page'

@app.route('/user/<username>')
def profile(username):
    return f'Hello, {username}'

if __name__ == '__main__':
    with app.test_request_context():
        print(url_for('index'))  # Output: /
        print(url_for('profile', username='Alice'))  # Output: /user/Alice
```

The `url_for()` function generates URLs based on the view function's name and any keyword arguments it requires. In this example, it produces the correct URLs for the `index` and `profile` routes.
````
